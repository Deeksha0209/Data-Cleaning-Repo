```python
#remove dollar sign from item_price col
df1['item_price'] = df1['item_price'].apply(lambda s:s.replace('$',''))
print(df1)
```

```python
#converting obj data type to float for item_price col

df1['item_price'] = df1['item_price'].astype(float)
df1.dtypes
```

```python
print(df1["item_name"].str.capitalize())
```

```python
#mode of choice desc col
df1.choice_description.mode()
```

```python
#drop rows containing na values from choice desc col
df1.dropna(subset=['choice_description'], inplace=True)
print(df1)
```

```python
#updated rows & col
df1.shape
```

```python
#0 null/missing values now
df1.isnull().sum()
```

```python
#sorting item_price col in asc order - this way we may analyse the Quantity wrt the price
df1.sort_values(by=['item_price'])
```

```python
#Grouping qty & item name wrt sum of price
df_new = pd.pivot_table(df1, index = ['item_name','quantity'], values = 'item_price', aggfunc={'sum'} )
print(df_new)
```

```python
#check for duplicated entries
df1.duplicated()
duplicate_rows = df1.duplicated(subset=['quantity', 'choice_description', 'item_name'])
print(duplicate_rows)
```

```python
#dropping duplicates from df
new_df = df1.drop_duplicates
print(new_df)
```
