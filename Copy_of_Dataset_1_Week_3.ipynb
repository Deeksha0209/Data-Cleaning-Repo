```python
import numpy as np
import pandas as pd
import io
df = pd.read_csv(io.BytesIO(uploaded['Data-cleaning-for-beginners-using-pandas(1).csv']))
print(df)
```

```python
#dropping Index column from the dataset because it is extra.
new_df1 = df.drop(columns=["Index"])
print(new_df1)
```

```python
#gives datatypes of each col
print(new_df1.dtypes)
```

```python
#gives no of rows & col
print(new_df1.shape)
```

```python
#extra step - for resetting index in a df
new_df1.reset_index(drop = True, inplace = True)
print(new_df1)
```

```python
#replace missing/NaN values in Age col with the mean of values in Age.
new_df1.Age.fillna(new_df1.Age.mean(), inplace = True)
print(new_df1)
```

```python
#converting Datatype float to Int in Age
new_df1['Age'] = new_df1['Age'].astype(int)
print(new_df1)
```

```python
#remove dollar sign & replace k with 000 from salary column using lambda method
new_df1['Salary'] = new_df1['Salary'].apply(lambda s:s.replace('$',''))
new_df1['Salary'] = new_df1['Salary'].apply(lambda t:t.replace('k','000'))
print(new_df1)
```

```python
#mean of rating col
new_df1.Rating.mean()
```

```python
#replace Nan entries in rating with mean & rounding off values in Rating
new_df1.Rating.fillna(new_df1.Rating.mean(), inplace = True)
new_df1['Rating']=new_df1['Rating'].round(decimals=1)
#Replacing -1.0 to mean values
new_df1['Rating'] = new_df1['Rating'].replace(-1.0,3.5)
print(new_df1)
```

```python
new_df1['Location'] = new_df1['Location'].apply(lambda l:l.replace(',',' '))
print(new_df1)
```

```python
#finding mode of Established column
new_df1.Established.mode()
#replacing -1 entries with mode of col
new_df1['Established'] = new_df1['Established'].replace(-1,1999)
print(new_df1)
```

```python
#replace -1 values to false in Easy Apply col
new_df1['Easy Apply'] = new_df1['Easy Apply'].str.replace('-1','FALSE')
print(new_df1)
```
